Applying the preprocessing methods to clean the data

```python
import string
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
def nlp_cleaner(sentence):
    punc_remover = [p for p in sentence if p not in string.punctuation]
    punc_join = ''.join(punc_remover)
    cleaner = [word for word in punc_join.split() if word.lower() not in stopwords.words('english')]
    return cleaner
```

```python
data_cleaned=data['tweet'].apply(nlp_cleaner)
```

Removed all punctuations from the sentences and extracted the words out of every sentences.

```python
data_cleaned.head(10)
```

Since Machine Learning Algorithm understands numbers. Lets convert the words into vectors based on its repeatation.

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer=nlp_cleaner, dtype='uint8')
tweet_vect = vectorizer.fit_transform(data['tweet']).toarray()
```

```python
x=tweet_vect
y=data['label']
print(x.shape,y.shape)
```

Applying Synthetic Minority Over Sampling to oversample the data and convert it into balanced data,so that the model performs better.

```python
from imblearn.over_sampling import SMOTE
over_sampler=SMOTE(random_state=2)
```

```python
res_x,res_y = over_sampler.fit_resample(x,y)
```

```python
print(res_x.shape,res_y.shape)
```

Train our model using Naive Bayes 

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(res_x,res_y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnbc= MultinomialNB()
mnbc.fit(xtrain,ytrain)
```

Check for Performance of Model

```python
from sklearn.metrics import classification_report, confusion_matrix
ypred=mnbc.predict(xtest)
cmx = confusion_matrix(ytest,ypred)
sns.heatmap(cmx,annot=True,cmap='mako',fmt='.2f')
```

```python
print(classification_report(ytest, ypred))
```
